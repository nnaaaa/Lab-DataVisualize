```python
import matplotlib.pyplot as plt
import numpy as np

# Tạo dữ liệu giả
labels = ['A', 'B', 'C', 'D']
data1 = [10, 20, 30, 40]
data2 = [5, 15, 25, 35]
data3 = [2, 4, 6, 8]
# Tạo biểu đồ
x = np.arange(len(labels))
fig, ax = plt.subplots()
ax.bar(x, data1, label='Data 1')
ax.bar(x, data2, bottom=data1, label='Data 2')
ax.bar(x, data3, bottom=np.array(data1) + np.array(data2), label='Data 3')

# Đặt tiêu đề và chú thích trục
ax.set_ylabel('Giá trị')
ax.set_xlabel('Nhãn')
ax.set_title('Biểu đồ chồng')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Hiển thị biểu đồ
plt.show()
print(x)
```
