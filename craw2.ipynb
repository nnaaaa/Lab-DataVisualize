```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Make a request to the website URL
url = 'https://www.worldometers.info/coronavirus/'
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find the elements you want to extract
table1 = soup.find('table', {'id': 'main_table_countries_today'})
table_body1 = table1.find('tbody')
table_header = table1.find('thead')
rows1 = table_body1.find_all('tr')

table2 = soup.find('table', {'id': 'main_table_countries_yesterday'})
table_body2 = table2.find('tbody')
rows2 = table_body2.find_all('tr')

table3 = soup.find('table', {'id': 'main_table_countries_yesterday2'})
table_body3 = table3.find('tbody')
rows3 = table_body3.find_all('tr')
```

```python
header = [ele.text.strip() for ele in table_header.find_all('th')]
```

```python
table_data1 = []
for row in rows1:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    table_data1.append(cols)

table_data2 = []
for row in rows2:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    table_data2.append(cols)

table_data3 = []
for row in rows3:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    table_data3.append(cols)
    
```

```python
df1 = pd.DataFrame(table_data1, columns=header)
df1.drop("#", inplace=True, axis=1)

df2 = pd.DataFrame(table_data2, columns=header)
df2.drop("#", inplace=True, axis=1)

df3 = pd.DataFrame(table_data3, columns=header)
df3.drop("#", inplace=True, axis=1)
df3
```

```python
df1.to_csv('./data/covid_3_6_2023.csv', index=False)
df2.to_csv('./data/covid_3_5_2023.csv', index=False)
df3.to_csv('./data/covid_3_4_2023.csv', index=False)
```
