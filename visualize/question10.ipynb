```python
import pandas as pd
import numpy as np
from pycountry_convert import country_name_to_country_alpha2
from geopy.geocoders import Nominatim
```

```python
def get_continent(col):
    try:
        cn_a2_code =  country_name_to_country_alpha2(col)
    except:
        cn_a2_code = 'Unknown'
    return cn_a2_code
```

```python
covid_df = pd.read_csv('../data/final_data.csv')
last_day = covid_df[covid_df['Date'] == '2023-03-10']
last_day['Country'] = last_day['Country'].apply(get_continent)
last_day.head()
```

```python
# rename continent name to continent code
continent_dict = {'North America': 'NA', 'South America': 'SA', 'Asia': 'AS', 'Europe': 'EU', 'Africa': 'AF', 'OC': 'Oceania', 'Unknown': 'Unknown'}
# replace continent name with continent code
last_day['Continent'] = last_day['Continent'].replace(continent_dict)
last_day
```

```python
# create a new column to store continent code and country code
last_day['Continent_Country'] = last_day['Continent'] + ',' + last_day['Country']
last_day.head()
```

```python
#function to get longitude and latitude data from country name
geolocator = Nominatim(user_agent="Phat-application")
def geolocate(country):
    try:
        # Geolocate the center of the country
        loc = geolocator.geocode(country)
        # And return latitude and longitude
        return (loc.latitude, loc.longitude)
    except:
        # Return missing value
        return np.nan
```

```python
last_day['Lat_Long'] = last_day['Country'].apply(geolocate)
last_day
```

```python
# save to file csv
last_day.to_csv('../data/last_day.csv', index=False)
```
