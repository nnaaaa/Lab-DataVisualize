# Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read data

```python
# Read the data
df = pd.read_csv('../data/final_data.csv')
df.head(10)
```

# Analytics for data visualization
### **Question 6. Is there a link between the number of tests and the number of cases?**
There are following steps to solve this problem:
1. Get the data to prepare visualization.
2. Create a scatter plot with number of tests on the x-axis and number of infections on the y-axis


#### **Step 1. Get the data to prepare visualization.**

```python
# Lấy dữ liệu từ dataframe đã có
data = df[['TotalCases', 'TotalTests']]
data.head(10)
```

#### **Step 2. Create a scatter plot with number of tests on the x-axis and number of infections on the y-axis.**

```python
# Vẽ scatter plot với TotalTests trục x, TotalCases trục y
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="TotalTests", y="TotalCases")

# Thêm regression plot
sns.regplot(data=data, x="TotalTests", y="TotalCases", scatter=False, color="red")

# Đặt tên cho trục x và y, và tiêu đề biểu đồ
plt.xlabel("Total Tests")
plt.ylabel("Total Cases")
plt.title("Relation between Total Tests and Total Cases")

# Hiển thị biểu đồ
plt.show()
```

    - Tuy nhiên, khi số lượng xét nghiệm tăng lên 1 mức độ nhất định thì số ca nhiễm lại không tăng nữa. Điều này có thể là do số lượng xét nghiệm đã đạt đến mức đủ để phát hiện ra các ca nhiễm mới.
