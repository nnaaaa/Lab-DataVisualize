# Analytics for data visualization

### **Question 8: Draw graph showing the number of new cases of all countries in 6 continents in 1 week?**
- There are many steps to answer this question:
    - Group by continent and day, then calculate the total number of new cases per day
    - Draw a graph of 6 lines each corresponding to each continent then comment

#### **Step 1. Group by continent and day, then calculate the total number of new cases per day.**

```python
covid_df = covid_df.loc[(covid_df['Date'] >= '2023-03-03') & (covid_df['Date'] <= '2023-03-10')]
covid_df.head(10)
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'])

continent_weekly = covid_df.groupby(['Continent', 'Date']).sum()
continent_weekly.head(10)
```

```python
#get continent, date, new cases columns
continent_weekly = continent_weekly['NewCases']
continent_weekly.head(20)
```

#### **Step 2. Draw a graph of 6 lines each corresponding to each continent then comment.**

```python
# Tạo figure và axes
fig, ax = plt.subplots(figsize=(12,6))

# Tạo một list chứa thông tin các châu lục và màu sắc tương ứng
continents = ['Asia', 'Europe', 'North America', 'South America', 'Africa', 'Oceania']
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown']

# Lặp qua từng châu lục và vẽ line plot
for i, continent in enumerate(continents):
    # Lọc ra các bản ghi có châu lục tương ứng
    continent_data = covid_df[covid_df['Continent'] == continent]
    # Lấy các giá trị NewCases của tuần gần đây nhất (để vẽ line cho biểu đồ)
    week_data = continent_data.groupby(['Country', 'Date'])['NewCases'].sum().reset_index().groupby('Date').sum().tail(7)
    # Tạo line plot với màu sắc tương ứng
    ax.plot(week_data['Date'], week_data['NewCases'], label=continent, color=colors[i])

# Thêm title, x-label, y-label, và legend
ax.set_title('New Cases by Continent for the Past Week')
ax.set_xlabel('Date')
ax.set_ylabel('New Cases')
ax.legend()

plt.show()
```
